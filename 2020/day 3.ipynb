```python
def count_trees(matrix,incy,incx):
    sx=matrix.shape[1]
    sy=matrix.shape[0]
    x=0
    y=0

    arboles=0
    continuar=True
    while continuar:
        y+=incy
        x=(x+incx)%(sx)
        if y>=sy:
            continuar=False
        else:
            if (matrix[y,x]==1):
                arboles+=1
        
    return arboles
    
```
