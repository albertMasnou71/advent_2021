```python
def grow_image_diagonal(basin):
    new=(basin.copy()>0).astype(int)*0
    new[:-1,:]=new[:-1,:]+basin[1:,:]
    new[1:,:]=new[1:,:]+basin[:-1,:]
    new[:,:-1]=new[:,:-1]+basin[:,1:]
    new[:,1:]=new[:,1:]+basin[:,:-1]
    
    new[:-1,:-1]=new[:-1,:-1]+basin[1:,1:]
    new[:-1,1:]=new[:-1,1:]+basin[1:,:-1]
    new[1:,1:]=new[1:,1:]+basin[:-1,:-1]
    new[1:,:-1]=new[1:,:-1]+basin[:-1,1:]
   
    
    
    
    
    return new
```

```python
def min_local_energy(min_energy,energy_map):
    new=min_energy.copy()
    new[:-1,:]=np.minimum(new[:-1,:],min_energy[1:,:])
    new[1:,:]=np.minimum(new[1:,:],min_energy[:-1,:])
    new[:,:-1]=np.minimum(new[:,:-1],min_energy[:,1:])
    new[:,1:]=np.minimum(new[:,1:],min_energy[:,:-1])
    new=new+energy_map
    new=np.minimum(min_energy,new)
    return new
```
