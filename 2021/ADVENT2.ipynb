```python
def processar_linea(linea):
    pila=[]
    continuar=True
    contra={'[':']',
            '(':')',
            '{':'}',
            '<':'>',
            }

    errors={']':57,
            ')':3,
            '}':1197,
            '>':25137,
            }
    i=0
    while continuar:
       # print(pila)
        if i<len(linea):
            if linea[i] in '[(<{':
                pila.append(linea[i])
            elif linea[i] in ']})>':
                sortida=pila.pop()
                if linea[i]!=contra[sortida]:
                    error=errors[linea[i]]
                    continuar=False
                    return error,[]
            i+=1
        else:
            continuar=False
            return 0
```

```python
def processar_linea_2(linea):
    pila=[]
    continuar=True
    contra={'[':']',
            '(':')',
            '{':'}',
            '<':'>',
            }

    errors={']':57,
            ')':3,
            '}':1197,
            '>':25137,
            }
    score_char={'[':2,
            '(':1,
            '{':3,
            '<':4,
            }

    
    i=0
    while continuar:
       # print(pila)
        if i<len(linea):
            if linea[i] in '[(<{':
                pila.append(linea[i])
            elif linea[i] in ']})>':
                sortida=pila.pop()
                if linea[i]!=contra[sortida]:
                    error=errors[linea[i]]
                    continuar=False
                    return 0
            i+=1
        else:
            continuar=False
            score=0
            while len(pila):
                score=score*5
                score+=score_char[pila.pop()]
            return score
```

```python
def grow_image(basin):
    new=(basin.copy()>0).astype(int)
    new[:-1,:]=new[:-1,:]+basin[1:,:]
    new[1:,:]=new[1:,:]+basin[:-1,:]
    new[:,:-1]=new[:,:-1]+basin[:,1:]
    new[:,1:]=new[:,1:]+basin[:,:-1]
    
    new[:-1,:-1]=new[:-1,:-1]+basin[1:,1:]
    new[:-1,1:]=new[:-1,1:]+basin[1:,:-1]
    new[1:,1:]=new[1:,1:]+basin[:-1,:-1]
    new[1:,:-1]=new[1:,:-1]+basin[:-1,1:]
    new+=basin
    
    
    
    
    return new
```

```python
matrix=np.array([np.array([int(y) for y in x]) for x in txt.split('\n')])
```

```python
def evolve (matrix_in):
    matrix=matrix_in.copy()
    matrix+=1
    activats_nous=(matrix>9).astype(int)
    flashes=activats_nous
    continuar=flashes.sum()>0
    while continuar:
        estimul=grow_image(activats_nous) #*(1-activats)
        matrix+=estimul

        activats=(matrix>9).astype(int)
        
        activats_nous=activats * (1-flashes)
        
        flashes=((activats+flashes)>0).astype(int)
        
        
        if activats_nous.sum()==0:
            continuar=False
    matrix=((matrix<10)*matrix)
    return matrix
```
