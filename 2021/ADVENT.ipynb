```python
def is_bingo(matrius_in,elegit):
    r=None
    r_max=None
    max_i=None
    for i in range(len(matrius_in)):
        alpha=(matrius_in[i]==elegit)
        before=max((matrius_in[i]<0).sum(axis=0).max(),(matrius_in[i]<0).sum(axis=1).max())==5
        
        matrius_in[i]=((alpha*-1))+((1-alpha)*matrius_in[i])
        after=max((matrius_in[i]<0).sum(axis=0).max(),(matrius_in[i]<0).sum(axis=1).max())==5
        
        if after and (not before):
            unmarked=(matrius_in[i]*(matrius_in[i]>0)).sum()
            r=unmarked*elegit
            r_max=matrius_in[i].copy()
            max_i=i
    return r,r_max,max_i

       
```
