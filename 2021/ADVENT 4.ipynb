```python
def generar_estat(sequence):
    estat={}
    for x,y in zip(list(sequence)[:-1],list(sequence)[1:]):
        k=x+y
        if k in estat.keys():
            estat[k]=estat[k]+1
        else:
            estat[k]=1
    return estat
```

```python
    
    
def actualitzar_estats(estat,traductor):
    nou_estat=estat.copy()
    for k in traductor.keys():
        if estat.get(k,0)>0:
            nou_estat[k]=nou_estat[k]-estat.get(k,0)
            nou_estat[traductor[k][0]]=nou_estat.get(traductor[k][0],0)+estat.get(k,0)
            nou_estat[traductor[k][1]]=nou_estat.get(traductor[k][1],0)+estat.get(k,0)
    return nou_estat
def contar_lletres(estat):
    lletres={}
    for k in estat.keys():
        lletres[k[0]]=lletres.get(k[0],0)+estat[k]
        lletres[k[1]]=lletres.get(k[1],0)+estat[k]
    for k in lletres.keys():
        lletres[k]=(lletres[k]+1)//2
    return lletres
def rati_lletras(d):
    dd=[d[k] for k in d.keys()]
    return max(dd)-min(dd)
```

```python
def expandir (seq,traductor):
    out=seq[0]
    for i in range(1,len(seq)):
        insertion=traductor.get(seq[i-1:i+1],'')
        out+=insertion
        out+=seq[i]
    return out
```

```python
def contar(sequence):
    caracteres=list(np.unique(list(sequence)))
    cuentas=[sum([1 for x in list(sequence) if x==y]) for y in caracteres]
    return {'caracter':caracteres,
           'n':cuentas}
```
