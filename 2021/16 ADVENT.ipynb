```python
def parse_package(binary_string):
    #print(f'parse: {binary_string}')
    messaje={'packet_version':to_decimal(binary_string[:3]),
            'type_id':to_decimal(binary_string[3:6]),
            }
    
    remaining_bits=binary_string[6:]
    messaje['remaining_bits']=remaining_bits
    #print(messaje)
    if messaje['type_id']==4:
        messaje['type']='literal value'
        #aixo igual no cal
        '''      
        while ceil(len(remaining_bits)/4)!=(len(remaining_bits)/4)
            remaining_bits+='0'
        '''
        literal_value=''
        while remaining_bits[0]=='1':
            literal_value+=remaining_bits[1:5]
            remaining_bits=remaining_bits[5:]
        literal_value+=remaining_bits[1:5]
        remaining_bits=remaining_bits[5:]
        messaje['value_bin']=literal_value
        messaje['value']=to_decimal(literal_value)
    else:
        messaje['type']='operator'
        messaje['package_lenght_type_id']=remaining_bits[0]
        remaining_bits=remaining_bits[1:]
        if messaje['package_lenght_type_id']=='0':
          
            messaje['package_total_lenght']=to_decimal(remaining_bits[:15])
            #print(messaje)
            remaining_bits=remaining_bits[15:]
            pending_bits=remaining_bits[:messaje['package_total_lenght']]
            messaje['content']=[]
            while len(pending_bits):
                
                content,pending_bits=parse_package(pending_bits)
                messaje['content'].append(content.copy())
            remaining_bits=remaining_bits[messaje['package_total_lenght']:]
        else:
            messaje['number_of_packets']=to_decimal(remaining_bits[:11])
            #print(messaje)
            remaining_bits=remaining_bits[11:]
            messaje['content']=[]
            for i in range(messaje['number_of_packets']):
                r,remaining_bits=parse_package(remaining_bits)
                messaje['content'].append(r)
            
        messaje['remaining_bits']=remaining_bits
        #print(messaje)
    return messaje,remaining_bits
    
```

```python
operacions={0:'suma',
           1:'producte',
           2:'el_minim',
           3:'el_maxim',
           4:'valor',
           5:'mes_gran',
           6:'mes_petit',
           7:'iguals'}
```

```python
def messaje_to_operation(messaje):
    print(f'processing: {messaje}')
    if messaje['type_id']==4:
        operation=f'{messaje["value"]}'
    else:
        operation=operacions[messaje['type_id']]+' ( ['
        operation+=messaje_to_operation(messaje['content'][0])
        for c in messaje['content'][1:]:
            operation+=' , '+ messaje_to_operation(c)
        operation+='])'
    return operation
    
```

```python
def suma(l):
    return sum(l)

def producte(l):
    ret_val=1
    for x in l:
        ret_val=ret_val*x
    return ret_val

def el_minim(l):
    return min(l)

def el_maxim(l):
    return max(l)

def mes_gran(l):
    return (l[0]>l[1])*1

def mes_petit(l):
    return (l[0]<l[1])*1

def iguals(l):
    return (l[0]==l[1])*1

        
```
