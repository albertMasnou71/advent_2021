```python
def read_scannners(txt):
    lines=txt.split('\n')
    scanner_lines=[i for i,x in enumerate(lines) if x.startswith('---')]
    scanner_lines+=[len(lines)+1]
    scanner=[]
    for i in range(len (scanner_lines)-1): 
        scanner.append(np.array([np.array([int(y) for y in x.split(',')]) for x in lines[scanner_lines[i]+1:scanner_lines[i+1]-1]]))
    return scanner
```

```python
# generar triangles distancies entre punts
def generate_invariants(scanner):
    triangles=[]
    for i in range(len(scanner)):
        triangle={}
        for i1,e1 in enumerate(scanner[i]):
            for i2,e2 in enumerate(scanner [i]):
                for i3,e3 in enumerate(scanner [i]):
                    if i1!=i2 and i2!=i3:
                        d1=((e1-e2)**2).sum()**0.5
                        d2=((e1-e3)**2).sum()**0.5
                        d3=((e2-e3)**2).sum()**0.5
                        s=sorted([d1,d2,d3])
                        h=hash(str(s))

                        triangle[h]=triangle.get(h,[])+[i1,i2,i3]
        triangles.append(triangle)
    return triangles
            
```

```python
# determinar equivalencias
def determinar_equivalencias (i,j,scanner,triangles):
    comuns=list(set(triangles[i].keys()).intersection(triangles[j].keys()))
    votes=np.zeros((len(scanner[i]),len(scanner[j])))
    for c in comuns:
        for x_i in triangles[i][c]:
            for x_j in triangles[j][c]:
                   votes[x_i,x_j]+=1
    equivalencias=[]
    for i in range(votes.shape[0]):
        if max(votes[i,:])>0:
            equivalencias.append([i,np.argmax(votes[i,:]),max(votes[i,:])])
    return equivalencias
```

```python
def construct_matrix_of_equivalences(scanner,triangles):
    d=np.zeros((len(scanner),len(scanner)))
    equivalences=[]
    for i in range (len(scanner)):
        column=[]
        for j in range (len(scanner)):
            e=(determinar_equivalencias(i,j,scanner,triangles))
            d[i,j]=len(e)
            column.append(e)
        equivalences.append(column)
    return d,equivalences
    
```

```python
def determine_resolution_order(overlapping):
    selected=[0]
    pending=[x for x in range(1,len(overlapping))]
    resolution_order=[]
    d=overlapping.copy()
    while len(pending):
        print(selected)
        d[:,selected]=0
        if len(selected)==1:
            origin=selected[0]
            destine=d[origin,:].argmax()
        else:
            selected_winner=d[selected,:].max(axis=1).argmax()
            origin=selected[selected_winner]
            destine=d[origin,:].argmax()
        selected.append(destine)
        pending.remove(destine)
        resolution_order.append([origin,destine])
    return resolution_order
```

```python

def compute_transformation(origen,destino,equivalences,scanner):
    '''transformation to get origen in destino coordinates'''
    i=destino
    j=origen
    
    
    actual_equs=equivalences[i][j]
    actual_equs=[x for x in actual_equs if x[2]>2000]
    origin_points=np.zeros((len(actual_equs),3))
    destination_points=np.zeros((len(actual_equs),3))
    for k,e in enumerate(actual_equs):
        origin_points[k,:]=scanner[i][e[0],:]
        destination_points[k,:]=scanner[j][e[1],:]
    o=origin_points-origin_points.mean(axis=0)
    o=o/((o**2).sum(axis=0))**0.5
    d=destination_points-destination_points.mean(axis=0)
    d=d/((d**2).sum(axis=0))**0.5
    rotations=np.zeros((3,3))
    for i in range(3):
        for j in range (3):
            c=np.matmul(o[:,i],d[:,j].T)
            if (abs(abs(c)-1)<0.01):
                rotations[i,j]=np.sign(c)

    destination_rotated=np.matmul(destination_points,rotations.T)
    shift=np.mean((destination_rotated-origin_points),axis=0).astype(int)
    errors=np.std((destination_rotated-origin_points),axis=0)
    points=np.zeros((len(scanner[origen]),3))
    for n_point in range(len(scanner[origen])):
        points[n_point,:]=scanner[origen][n_point]
    
    transformed_points=apply_transformation (points,rotations,shift).astype(int)
    return transformed_points,rotations,shift,errors
```

```python
def apply_transformation (points,rotations,shift):
    rotated=np.matmul(points,rotations.T)
    transformed_points=rotated-shift
    return transformed_points.astype(int)
```
