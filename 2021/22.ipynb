```python
def intersection(min_x,max_x,minimo,maximo):
    if (min_x>maximo) or (minimo>max_x):
        intersect=False
        r=[]
        d=0
    else:
        intersect=True
        r=[max(min_x,minimo),min(max_x,maximo)]
        d=r[1]-r[0]
    return intersect,d,r
```

```python
def box_intersection (box1,box2):
    ix,dx,rx=intersection(box1['x'][0],box1['x'][1],box2['x'][0],box2['x'][1])
    iy,dy,ry=intersection(box1['y'][0],box1['y'][1],box2['y'][0],box2['y'][1])
    iz,dz,rz=intersection(box1['z'][0],box1['z'][1],box2['z'][0],box2['z'][1])
    intersection_box={'x':rx,'y':ry,'z':rz}
    return ix*iy*iz,dx*dy*dz,intersection_box
    
    
    
```

```python
def intervalos_interseccion(x,keep_x,exclude_intersection=True):
    '''trocea un rango en intervalos disjuntos partiendolos con un intervalo interseccionado 
    de tal forma que en los intervalos no se modificquen los intervalos definidos por keep_x
    '''
    
    intervalos=[]
    if x[0]<keep_x[0]:
        intervalos+=[[x[0],keep_x[0]-1]]
    if not exclude_intersection:
        intervalos+=[keep_x]
    if x[1]>keep_x[1]:
        intervalos+=[[keep_x[1]+1,x[1]]]
        
    
        
    return intervalos
    
```

```python
def split_box(box,keep_box):
    ixs=intervalos_interseccion(box['x'],keep_box['x'],False)
    iys=intervalos_interseccion(box['y'],keep_box['y'],False)
    izs=intervalos_interseccion(box['z'],keep_box['z'],False)
    
    result=[]
    for x in ixs:
        for y in iys:
            for z in izs:
                
                new_box={'x':x,'y':y,'z':z}
                if not is_equal(new_box,keep_box):
                    result.append(new_box)
                
    return result
                        
    
```

```python
def is_equal(box1,box2):
    return (box1['x']==box2['x']) & (box1['y']==box2['y']) & (box1['z']==box2['z'])

def contained (b_in,b_out):
    '''
    box2 contained in box1
    '''
    contained_x= ( (b_in['x'][0]>=b_out['x'][0]) & (b_in['x'][1]<=b_out['x'][1]))
    contained_y= ( (b_in['y'][0]>=b_out['y'][0]) & (b_in['y'][1]<=b_out['y'][1]))
    contained_z= ( (b_in['z'][0]>=b_out['z'][0]) & (b_in['z'][1]<=b_out['z'][1]))
    
    return contained_x & contained_y & contained_z
    
```

```python
def turn_on (list_disconected_boxes,box):
    resulting_boxes=list_disconected_boxes.copy()
    
    pending_insertion_boxes=[box]
    while len(pending_insertion_boxes):
        actual_box=pending_insertion_boxes.pop()
        continuar=len(list_disconected_boxes)>0
        intersect=False
        i=0
        while continuar:
            
            list_box=list_disconected_boxes[i]
            
            intersect,d,intersection_box=box_intersection (list_box,actual_box)
            if intersect:
                if not contained(actual_box,list_box):
                    new_boxes=split_box(actual_box,intersection_box)
                    pending_insertion_boxes+=new_boxes
                continuar=False
            else:
                i+=1
                if i>=len(list_disconected_boxes):
                    continuar=False
                
            
        if not intersect:
            resulting_boxes.append(actual_box.copy())
            
    return resulting_boxes
                

    
```

```python
def turn_off (list_disconected_boxes,box):
    
    resulting_boxes=[]
    
    for on_box in list_disconected_boxes:
        intersect,d,intersection_box=box_intersection (on_box,box)
        if intersect:
            if  contained(on_box,box):
            else:
                new_boxes=split_box(on_box,intersection_box)
                resulting_boxes+=new_boxes
                
        else:
            resulting_boxes.append(on_box)
    
    return resulting_boxes
```
