```python
def grow_image_diagonal(basin):
    new=(basin.copy()>0).astype(int)
    new[:-1,:]=new[:-1,:]+basin[1:,:]
    new[1:,:]=new[1:,:]+basin[:-1,:]
    new[:,:-1]=new[:,:-1]+basin[:,1:]
    new[:,1:]=new[:,1:]+basin[:,:-1]
    
    new[:-1,:-1]=new[:-1,:-1]+basin[1:,1:]
    new[:-1,1:]=new[:-1,1:]+basin[1:,:-1]
    new[1:,1:]=new[1:,1:]+basin[:-1,:-1]
    new[1:,:-1]=new[1:,:-1]+basin[:-1,1:]
    new+=basin
    
    
    
    
    return new
```

```python

def min_local_energy(min_energy,energy_map):
    new=min_energy.copy()
    new[:-1,:]=np.minimum(new[:-1,:],min_energy[1:,:])
    new[1:,:]=np.minimum(new[1:,:],min_energy[:-1,:])
    new[:,:-1]=np.minimum(new[:,:-1],min_energy[:,1:])
    new[:,1:]=np.minimum(new[:,1:],min_energy[:,:-1])
    new=new+energy_map
    new=np.minimum(min_energy,new)
    return new
```

```python
def replicar_mapa(energia):
    c=energia
    anterior=energia
    for i in range(4):
        new=anterior+1
        alpha=new<=9
        new=(new*alpha)+(1-alpha)
        anterior=new
        c=np.concatenate((c,new),axis=1)
    fila=c   
    anterior=c
    for i in range(4):
        new=anterior+1
        alpha=new<=9
        new=(new*alpha)+(1-alpha)
        anterior=new
        fila=np.concatenate((fila,new),axis=0)
    return fila
```
