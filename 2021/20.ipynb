```python
lineas=txt.split('\n')
separador=[i for i,x in enumerate(lineas) if len(x)==0][0]
algoritmo=[1*(x=='#') for x in list(''.join(lineas[:separador]))]
imagen=np.array([np.array([(y=='#')*1 for y in list(x)]) for x in lineas[separador+1:]])
```

```python
iteraciones=50
```

```python
sx=imagen.shape[1]
sy=imagen.shape[0]
marco=iteraciones
infinite_image=np.zeros((sy+((iteraciones)*6),sx+((iteraciones)*6)))
infinite_image[(iteraciones*3):(iteraciones*3)+sy,(iteraciones*3):(iteraciones*3)+sx]=imagen
plt.figure()
plt.imshow(infinite_image)
for i in tqdm(range(iteraciones)):
    output_image=infinite_image*0
    for x in range(1,infinite_image.shape[1]-2):
        for y in range(1,infinite_image.shape[0]-2):
            code=(infinite_image[y-1:y+2,x-1:x+2].ravel()*(1<<(8-np.arange(9)))).sum().astype(int)

            output_image[y,x]=algoritmo[code]
    output_image[:4,:]=0
    output_image[-3:,:]=0
    output_image[:,:4]=0
    output_image[:,-3:]=0
    
    
    
    #plt.figure()
    #plt.imshow(output_image)
    infinite_image=output_image
```
