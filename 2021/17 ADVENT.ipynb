```python
def simula_velocitat(x,y,dx,dy,min_x,max_x,min_y,max_y):
    intercept=False
    nx=x
    ny=y
    sequence=[[nx,ny]]
    speed_x=dx
    speed_y=dy
    continuar=True
    if (nx>=min_x) and (nx<=max_x) and (ny>=min_y) and (ny<=max_y):
        intercept=True
        continuar=False
    if ny<min_y:
        continuar=False
            
    while continuar:
        nx=nx+speed_x
        ny=ny+speed_y
        speed_x=speed_x-np.sign(speed_x)
        speed_y=speed_y-1
        sequence.append([nx,ny])
        if (nx>=min_x) and (nx<=max_x) and (ny>=min_y) and (ny<=max_y):
            intercept=True
            continuar=False
        if (ny<min_y) or (x>max_x):
            continuar=False
    return intercept,sequence
    
        
        
    
```

```python
intercept,trajectory=simula_velocitat(0,0,6,3,20,30,-10,-5)
plt.figure()
plt.plot([x[0] for x in trajectory],[x[1] for x in trajectory],'og' if intercept else 'or')
```
